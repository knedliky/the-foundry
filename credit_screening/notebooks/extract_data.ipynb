# Extract Data

Japanese Credit Screening data was sourced from the UC Irvine Machine Learning Repository.

Includes domain theory (generated by talking to Japanese domain experts); data in Lisp.

## Dataset Information
### Additional Information

Examples represent positive and negative instances of people who were and were not  granted credit.

The theory was generated by talking to the individuals at a Japanese company that grants credit.


```python
import zipfile
import requests
from io import BytesIO 
```

```python
url = 'https://archive.ics.uci.edu/static/public/28/japanese+credit+screening.zip'
r = requests.get(url)

if r.status_code in range(200,300):
    data = BytesIO(r.content)

    with zipfile.ZipFile(data, 'r') as z:
        z.extractall('../data/bronze')
```
